```python
start_date="2015-01-01"
end_date="2019-05-07"
```

```python
def sum(f):
    data=yf.download(f,start=start_date,end=end_date)
    df=data["Close"]
    price = df.squeeze()
    
    company = Stocker(price)
    model, model_data = company.create_prophet_model(days=1)
    
    gprice=model_data["yhat"]
    reference=price[len(price)-1]
    guess=gprice[len(gprice)-1]
    
    return guess,reference
```
